###**What’s The Difference Between The Python append() and extend() Methods?**

**Append()-**

1. The argument element given as input is appended to the end of the list.

2. The list's length grows by one.

3. An iterable passed as an argument appends without any change as a single element to the end of the list.

5. Has a constant time complexity of O(1)

**Extend-**

1. Each element of the iterable passed as an argument gets appended to the end of the list

2. Length of the list increases by the number of elements in the iterable

3. An iterable passed as an argument will append each of its elements to the end of the list

4. Has a time complexity of O(k) where k is the length of the iterable.

###**Write a Python program to get the frequency of the elements in a list.**

```python
list1 = ['1', 'A2', 'B', 'C2', 'B3', 'D', 'D', 'A', 'B']
frequency = {}

# iterating over the list
for item in list1:
   # checking the element in dictionary
   if item in frequency:
      # incrementing the counr
      frequency[item] += 1
   else:
      # initializing the count
      frequency[item] = 1

# printing the frequency
print(frequency)
```

###**Write a Python program to generate all sublists of a list**

```python
# function to generate all the sub lists
def sub_lists (l):
    lists = [[]]
    for i in range(len(l) + 1):
        for j in range(i):
            lists.append(l[j: i])
    return lists
 
# driver code
l1 = [1, 2, 3]
print(sub_lists(l1))
```

###**Write a Python program to find common items from two lists**

```python

def common_member(a, b):
    result = [i for i in a if i in b]
    return result
 
a = [1, 2, 3, 4, 5]
b = [5, 6, 7, 8, 9]
 
print("The common elements in the two lists are: ")
print(common_member(a, b))
```

###**How to flatten a list in python?**

```python
#Flatten List using Nested for Loop
nestedlist = [ [1, 2, 3, 4], ["Ten", "Twenty", "Thirty"], [1.1,  1.0E1, 1+2j, 20.55, 3.142]]
flatlist=[]
for sublist in nestedlist:
    for element in sublist:
        flatlist.append(element)
print(flatlist)


#Flatten List using List Comprehension
nestedlist = [ [1, 2, 3, 4], ["Ten", "Twenty", "Thirty"], [1.1,  1.0E1, 1+2j, 20.55, 3.142]]
flatlist=[element for sublist in nestedlist for element in sublist]
print(flatlist)


```

###**How to sort a list in ascending and descending order without using the sort**

```python
my_list = [4, 2, 3, -1, -2, 0, 1]

for i in range(len(my_list)):
    for j in range(i + 1, len(my_list)):

        if my_list[i] > my_list[j]:
            my_list[i], my_list[j] = my_list[j], my_list[i]

print(my_list)

#for decending order
my_list.reverse()

print(my_list)
```

###**33. Difference between indexing and Slicing**

**Indexing-**

1. Indexing starts from 0. Index 0 represents the first element in the sequence.

2. Negative indexing starts from -1. Index -1 represents the last element in the sequence.

3. Attempting to use an index that is too large will result in an IndexError.

4. Assigning to an indexed position for immutable objects like strings, tuple, range object will raise a TypeError.

**slicing**
1. We can specify a range of indexes

2. s[1:3] — Returns element from the first index to the third index(excluded).

3. Slice indices have useful defaults; an omitted first index defaults to zero, an omitted second index defaults to the size of the string being sliced.

4. 
s[1:5:2]-Returns element from index 1 till index 5(excluded) using step 2.

5. Out of range indexes are handled gracefully when used for slicing.

6. While assignment to slicing, no need to specify the same number of elements. We can assign any number of elements. This will even change the size of the list.

###**Difference between copy and deep copy**

**Shallow copy-**
1. A shallow copy create new object wich store refrance of original elements
2. Shallow copy dosen't create a copy of nested objects
3. Shallow copy is faster as compaired to deep copy
4. Shallow copy reflects changes made to the copied object in original object 
5. syntax - copy.copy(x)

**Deep Copy-**
1. A deep copy creates new object and recursively adds the copies of nested object present in the original element
2. Deep copy creates copy of original object and its nested object
3. Deep copy is slower comparatively 
4. Deep copy dosen't reflect changes made to copied object in the original object
5. syntax - copy.deepcopy(x)
